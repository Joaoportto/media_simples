```python
import pandas as pd
import matplotlib.pyplot as plt
tabela_df = pd.read_csv("student_marks.csv")
tabela_df['Média'] = tabela_df.loc[:, 'Test_1':'Test_12'].mean(axis=1)
media = tabela_df['Média'].round(0)
plt.figure(figsize = ( 10 , 6))
plt.bar(media.index, media.values, color ="blue")
plt.xlabel('Alunos')
plt.ylabel( 'Média')
plt.title("Média das notas dos alunos")
plt.xticks(rotation=  45)
plt.tight_layout()
plt.show()

```
